```python
# –ù–∏–∂–µ –ø—Ä–∏–≤–µ–¥–µ–Ω—ã –¥–∏–∞–º–µ—Ç—Ä—ã –∫–æ—Ä–æ–Ω–∞—Ä–Ω—ã—Ö –∞—Ä—Ç–µ—Ä–∏–π –ø–æ—Å–ª–µ –ø—Ä–∏–µ–º–∞ –Ω–∏—Ñ–µ–¥–∏–ø–∏–Ω–∞ –∏ –ø–ª–∞—Ü–µ–±–æ. 
# –ü–æ–∑–≤–æ–ª—è—é—Ç –ª–∏ –ø—Ä–∏–≤–æ–¥–∏–º—ã–µ –Ω–∏–∂–µ –¥–∞–Ω–Ω—ã–µ —É—Ç–≤–µ—Ä–∂–¥–∞—Ç—å, —á—Ç–æ –Ω–∏—Ñ–µ–¥–∏–ø–∏–Ω –≤–ª–∏—è–µ—Ç –Ω–∞ –¥–∏–∞–º–µ—Ç—Ä –∫–æ—Ä–æ–Ω–∞—Ä–Ω—ã—Ö –∞—Ä—Ç–µ—Ä–∏–π?
```

```python
import numpy as np
import scipy.stats as stats

x = np.array([2.5, 2.2, 2.6, 2, 2.1, 1.8, 2.4, 2.3, 2.7, 2.7, 1.9])
y = np.array([2.5, 1.7, 1.5, 2.5, 1.4, 1.9, 2.3, 2.0, 2.6, 2.3, 2.2])

stats.ttest_ind(x, y)    
```

```python
print(f'–¢–∞–±–ª–∏—á–Ω—ã–π –∫—Ä–∏—Ç–µ—Ä–∏–π = {stats.t.ppf(0.975, 22-2)}')
```

```python
x1 = np.mean(x)
y1 = np.mean(y)
t = (x1 - y1)/np.sqrt(np.var(x, ddof=1)/len(x) + np.var(y, ddof=1)/len(y))
print(t)
```

```python
# t=1.3283 < t—Ç=2.085 => –ø—Ä–∏–Ω–∏–º–∞–µ–º 0-—é –≥–∏–ø–æ—Ç–µ–∑—É
```

```python
# –ù–∏–∂–µ –ø—Ä–∏–≤–µ–¥–µ–Ω—ã –¥–∞–Ω–Ω—ã–µ –∏–∑ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è –§—Ä–µ–±–∞ –∏ –£–∞–π—Ç–∞, –ø–æ—Å–≤—è—â–µ–Ω–Ω–æ–º—É –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—é
# —Å–æ—Å—Ç–æ—è–Ω–∏—é –ª–µ–≥–∫–∏—Ö. . –ú—ã –≤–æ–∑—å–º–µ–º –¥–∞–Ω–Ω—ã–µ –¥–ª—è –≥—Ä—É–ø–ø—ã –ª—é–¥–µ–π, –∫–æ—Ç–æ—Ä—ã–µ —Ä–∞–±–æ—Ç–∞–ª–∏ –≤ –Ω–∞–∫—É—Ä–µ–Ω–Ω–æ–º
# –ø–æ–º–µ—â–µ–Ω–∏–∏ –∏ –¥–ª—è –ª—é–¥–µ–π, –≤—ã–∫—É—Ä–∏–≤–∞—é—â–∏—Ö –Ω–µ–±–æ–ª—å—à–æ–µ —á–∏—Å–ª–æ —Å–∏–≥–∞—Ä–µ—Ç –≤ –¥–µ–Ω—å.
# –û–±—ä–µ–º—ã –≤—ã–±–æ—Ä–æ–∫ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ ‚Äì –ø–æ 200 —á–µ–ª–æ–≤–µ–∫. –î–ª—è –ª—é–¥–µ–π, —Ä–∞–±–æ—Ç–∞—é—â–∏—Ö –≤ –Ω–∞–∫—É—Ä–µ–Ω–Ω–æ–º
# –ø–æ–º–µ—â–µ–Ω–∏–∏ —Å—Ä–µ–¥–Ω—è—è —Å–∫–æ—Ä–æ—Å—Ç—å —Å—Ä–µ–¥–∏–Ω—ã –≤—ã–¥–æ—Ö–∞ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç 2,72, std = 0.71, –∞ –≤—ã–∫—É—Ä–∏–≤–∞—é—â–∏—Ö
# –Ω–µ–±–æ–ª—å—à–æ–µ —á–∏—Å–ª–æ —Å–∏–≥–∞—Ä–µ—Ç 2,63, std = 0.73.–ú–æ–∂–Ω–æ –ª–∏ —Å—á–∏—Ç–∞—Ç—å –æ–¥–∏–Ω–∞–∫–æ–≤–æ–π —Å—Ä–µ–¥–Ω—é—é —Å–∫–æ—Ä–æ—Å—Ç—å
# —Å–µ—Ä–µ–¥–∏–Ω—ã –≤—ã–¥–æ—Ö–∞ –æ–¥–∏–Ω–∞–∫–æ–≤–æ–π –≤ –æ–±–µ–∏—Ö –≥—Ä—É–ø–ø–∞—Ö?
```

```python
# –ù0: Œº1 = Œº2
# –ù1: Œº1 != Œº2 –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π –∫—Ä–∏—Ç–µ—Ä–∏–π

n = 200
x1 = 2.72
x2 = 2.63
s1 = 0.71
s2 = 0.73

# –†–∞—Å—á–µ—Ç –∫—Ä–∏—Ç–µ—Ä–∏—è –°—Ç—å—é–¥–µ–Ω—Ç–∞ t
th = (x1-x2)/((s1**2 + s1**2)/n)**0.5
print(t)
```

```python
print(f'–¢–∞–±–ª–∏—á–Ω—ã–π –∫—Ä–∏—Ç–µ—Ä–∏–π = {stats.t.ppf(0.975, 400-2)}')
```

```python
# t–Ω=1.2499 < t—Ç=1.96 => –ø—Ä–∏–Ω–∏–º–∞–µ–º 0-—é –≥–∏–ø–æ—Ç–µ–∑—É
```

```python
# –£—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç—Å—è, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–π —Ä–æ—Å—Ç –º—É–∂—á–∏–Ω –Ω–∞—Ü–∏–æ–Ω–∞–ª—å–Ω–æ—Å—Ç–∏ –• 179,5.
# –ë—ã–ª–∞ –≤–∑—è—Ç–∞ –≤—ã–±–æ—Ä–∫–∞ –∏–∑ 100 —á–µ–ª–æ–≤–µ–∫, –ø–æ –∫–æ—Ç–æ—Ä–æ–π –ø–æ–ª—É—á–∏–ª–æ—Å—å —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ 178,5. 
# –ü—Ä–æ–≤–µ—Ä–∏—Ç—å —ç—Ç–æ —É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ —Å –ø–æ–º–æ—â—å—é –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–µ–≥–æ —Ç–µ—Å—Ç–∞, –µ—Å–ª–∏ –∏–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ 
# –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏ 3 —Å–º. –ê —É—Ä–æ–≤–µ–Ω—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –ø—Ä–∏–Ω—è—Ç—å –∑–∞ 1%
```

```python
# –§–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞–Ω–∏–µ –≥–∏–ø–æ—Ç–µ–∑:
# H0: Œº = Œº0 –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞
# H1: Œº < Œº0 –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞

# –í—ã–±–æ—Ä –∫—Ä–∏—Ç–µ—Ä–∏—è
# ùëç–Ω=(ùëã‚Äæ‚àí Œº)/(ùûº/‚àöùëõ)

n = 100
x = 178.5
mu = 179.5
sigma = 3

Zh = (x - mu)/(sigma/n**0.5)
print(Zh)
```

```python
# Œ±=0.01 (–ø–æ 0.005 —Å –¥–≤—É—Ö —Å—Ç–æ—Ä–æ–Ω)

print(f'–¢–∞–±–ª–∏—á–Ω—ã–π –∫—Ä–∏—Ç–µ—Ä–∏–π = {stats.norm.ppf(0.005)}')
```
